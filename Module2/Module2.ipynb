```python
def get_rating(user_ratings, user, movie):
    return user_ratings.iloc[user][movie]

def average_rating(user_ratings, movie):
    summ = user_ratings[movie].sum(axis=0)
    return summ/(user_ratings.astype(bool)[movie].sum(axis=0))

def recommend_movies(users, movies, user, n_predictions):
    corr_movies = movies.set_index("movie_title").T.corr()
    #corr_movies = corr_movies.rename(columns={ corr_movies.columns[1]: "movie_title" })
    #corr_movies["movie_title"].keys().values
    rated_movies = users.iloc[user][users.iloc[user]!=0]
    unrated_movies = users.iloc[user][users.iloc[user]==0]
    
    keys = rated_movies.keys()[1:]
    relevant_movies = corr_movies[keys]
    
    predictions = {} 
    for i in unrated_movies.keys().values: 
        baseline = average_rating(users, i)
        related_movies = relevant_movies[relevant_movies.index == i]
        related_movies = related_movies[related_movies > 0.45].dropna(axis=1)
        summ = 0
        summ2 = 0
        for key, value in related_movies.iteritems(): 
            #print(get_rating(users, user, key))
            #print(average_rating(users, key))
            summ += value[0]*(get_rating(users, user, key) - average_rating(users, key)) 
            summ2 += value[0]
            
        if summ2 == 0:
            predictions[i] = 0
        else:
            predictions[i] = baseline + (summ / summ2)
    return sorted(predictions.items(), key=lambda x: x[1], reverse=True)[0:n_predictions]
```
